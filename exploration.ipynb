```python
import requests
import gzip
import shutil

# URL del archivo JSON de las "default cards"
url = "https://data.scryfall.io/default-cards/default-cards-20231017090458.json"
response = requests.get(url, stream=True)

# Comprobar si la respuesta es correcta
if response.status_code == 200:
    # Guardar el archivo
    with open("default-cards.json", "wb") as f:
        f.write(response.content)
else:
    print("Error al descargar el archivo.")
```

```python
import pandas as pd
import json
import numpy as np

# Cargar el archivo JSON
with open("default-cards.json", "r", encoding='utf-8') as f:
    data = json.load(f)

# Convertir el JSON a un DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
result = df.loc[df['name'] == 'Animar, Soul of Elements']
result
```

```python
filtered_df = df[['cmc', 'name', 'colors', 'keywords', 'loyalty', 'mana_cost', 'power', 'toughness', 'legalities', 'type_line', 'oracle_text', 'artist', 'rarity', 'set']]

filtered_df.head()
```

```python
import sqlite3

def convert_to_string(x):
    if isinstance(x, (list, dict)):
        return json.dumps(x)
    return x

for column in filtered_df.columns:
    filtered_df[column] = filtered_df[column].apply(convert_to_string)

conn = sqlite3.connect('cards.db')
filtered_df.to_sql('cards', conn, if_exists='replace', index=False)
```

```python
# Query para obtener todas las cartas legendarias de color verde, rojo y azul
query1 = """
SELECT name
FROM cards
WHERE type_line LIKE '%Legendary Creature%'
AND colors LIKE '["B", "G", "R", "U", "W"]'
"""
query2 = """
SELECT name
FROM cards
WHERE type_line LIKE '%Dragon%'
AND colors='["U"]'
AND NOT keywords LIKE '%Flying%'
"""

result = pd.read_sql_query(query1, conn)

result.drop_duplicates(inplace=True)
result
```
